# Test Functions

For testing the convergence behaviour of our reconstruction methods, we will use a number of test functions to test the behaviour.

```python
import sys
sys.path.append('../')

from gradient_reconstruction.util import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2,101)
y = np.linspace(0,2,101)
X,Y = np.meshgrid(x,y)
nodes = np.asarray([np.ravel(X), np.ravel(Y)]).T
```

## Constant Function

The constant function is given by
$$ f_{constant}(x,y) = a,$$
where $a$ is some number.

```python
a = 1
values = test_functions.constant_function(nodes, a)
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
surf = ax.plot_surface(X, Y, np.reshape(values, (-1,X.shape[1])))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

## Linear function

The linear function is given by
$$ f_{linear}(x,y) = a + b\ x + c\ y,$$
where $a,b$ and $c$ are some numbers.

```python
a = 1
b = -1
c = 2
values = test_functions.linear_function(nodes, a,b,c)
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
surf = ax.plot_surface(X, Y, np.reshape(values, (-1,X.shape[1])))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

## Quadratic Function

The linear function is given by
$$ f_{quadratic}(x,y) = a + b\ x + c\ y + d\ x^2 + e\ x \ y + f\ y^2,$$
where $a,b,c,d,e$ and $f$ are some numbers.

```python
a = 0.5
b = 0
c = 0
d = 1
e = 1
f = 1
values = test_functions.quadratic_function(nodes, a,b,c,d,e,f)
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
surf = ax.plot_surface(X, Y, np.reshape(values, (-1,X.shape[1])))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

## Sine

The sine function is given by
$$ f_{sine}(x,y) = a \sin(b(x+y)),$$
where $a$ and $b$ are some numbers.

```python
a = 1
b = 4
values = test_functions.sine_sum(nodes, a,b)
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
surf = ax.plot_surface(X, Y, np.reshape(values, (-1,X.shape[1])))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```
